```python
import numpy as np 
import matplotlib.pyplot as plt
from linalg import proj_vec 
```

```python
u = np.array([3,2])
v = np.array([1,1])
p = proj_vec(u, v)
```

```python
plt.quiver(0,0,u[0],u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(0,0,p[0],p[1], angles='xy', scale_units='xy', scale=1, color='g', label='p')
plt.xlim(-1,4); plt.ylim(-1,3)
plt.legend(); plt.grid(True); plt.axis('equal')
plt.title("Projection of u onto v")
plt.show()
```
